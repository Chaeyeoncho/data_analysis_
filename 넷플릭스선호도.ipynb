# 국가별 인기 장르 집계

```python
# 관련 컬럼 선택 및 결측치 제거
filtered_data = data[['country', 'listed_in']].dropna()

# 국가별로 데이터 분리 (국가명이 여러 개인 경우 분리)
filtered_data['country'] = filtered_data['country'].apply(lambda x: [country.strip() for country in x.split(',')])
filtered_data = filtered_data.explode('country')

# 장르별로 데이터 분리 (여러 장르가 한 셀에 있는 경우 분리)
filtered_data['listed_in'] = filtered_data['listed_in'].apply(lambda x: [genre.strip() for genre in x.split(',')])
filtered_data = filtered_data.explode('listed_in')

# 유효한 국가 이름만 필터링 (영문자 및 공백만 포함된 문자열)
valid_countries = filtered_data['country'].str.contains('^[A-Za-z ]+$')
filtered_valid_countries = filtered_data[valid_countries]

# 각 국가별로 가장 인기 있는 장르 집계
popular_genre_by_country = filtered_valid_countries.groupby('country')['listed_in'].agg(
    lambda x: x.value_counts().idxmax())

# 결과 출력
print(popular_genre_by_country.head(20))
```

#국가별 장르 선호 빈도수 분석

```python

# 'country'와 'listed_in' 컬럼에 집중
data_filtered = netflix_data[['country', 'listed_in']].dropna()

# 'country' 컬럼 정제하기
data_filtered['country'] = data_filtered['country'].str.split(', ')
data_exploded_countries = data_filtered.explode('country')
data_exploded_countries['country'] = data_exploded_countries['country'].str.strip()
valid_countries = data_exploded_countries['country'].str.contains('^[a-zA-Z ]+$')
cleaned_data = data_exploded_countries[valid_countries]

# 'listed_in' 컬럼 처리하기
cleaned_data['listed_in'] = cleaned_data['listed_in'].str.split(', ')
cleaned_data_exploded = cleaned_data.explode('listed_in')

# 국가별로 'listed_in' 카테고리 빈도수 계산
country_genre_counts = cleaned_data_exploded.groupby(['country', 'listed_in']).size().unstack(fill_value=0)

# 결과 출력
print(country_genre_counts.head())  # 처음 몇 줄만 출력
```

#gdp, 장르, 국가 병합

```python
# 파일 경로 설정
genre_data_path = '/content/country_genre_counts.csv'
gdp_data_path = '/content/데분_ 2022년도 나라별 gdp .csv'

# 데이터 불러오기
genre_data = pd.read_csv(genre_data_path)
gdp_data = pd.read_csv(gdp_data_path)

# 장르 데이터에서 필요한 컬럼 선택 및 결측치 제거
genre_data_filtered = genre_data.melt(id_vars='country', var_name='listed_in', value_name='count')
genre_data_filtered = genre_data_filtered[genre_data_filtered['count'] > 0]

# GDP 데이터에서 필요한 컬럼 선택 및 컬럼 이름 변경
gdp_data_renamed = gdp_data.rename(columns={'Country': 'country', 'GDP 2022': 'gdp'}).dropna()

# 두 데이터셋 병합
merged_data = pd.merge(genre_data_filtered, gdp_data_renamed, on='country', how='left')

# 결과 정리: 각 국가별로 'country', 'listed_in', 'gdp' 형식으로 표현
merged_data['listed_in'] = merged_data['listed_in'].apply(lambda x: x.replace('_', ' ').title())
merged_data = merged_data[['country', 'listed_in', 'gdp']]

print(merged_data)

# 최종 데이터 저장
#output_path = '/content/drive/MyDrive/merged_data.csv'
#merged_data.to_csv(output_path, index=False)

#print("Data saved successfully to:", output_path)
```

## 회귀분석을 위한 범주형 데이터를 더미 변수로 변환

```python
# 필요한 컬럼만 선택
import pandas as pd
from sklearn.model_selection import train_test_split
data_filtered = data[['country', 'listed_in']].dropna()

# 'country'와 'listed_in' 컬럼에서 여러 항목이 포함된 경우 분리
data_filtered['country'] = data_filtered['country'].str.split(', ')
data_filtered = data_filtered.explode('country')
data_filtered['listed_in'] = data_filtered['listed_in'].str.split(', ')
data_filtered = data_filtered.explode('listed_in')

# 중복 데이터 제거
data_filtered.drop_duplicates(inplace=True)

# 더미 변수 생성
data_dummies = pd.get_dummies(data_filtered, columns=['country', 'listed_in'])
```

## Train data와 Test data 분할

```python
# 데이터를 훈련 세트와 테스트 세트로 분할 (80/20 비율)
train_set, test_set = train_test_split(data_dummies, test_size=0.2, random_state=42)

# 훈련 세트와 테스트 세트의 개요를 출력
train_set_info = train_set.info()
test_set_info = test_set.info()
train_set.head(), test_set.head()
```

## 회귀분석을 위한 다중공정성

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor


# VIF계산
vif_data = pd.DataFrame()
vif_data["Variable"] = data_dummies.columns
vif_data["VIF"] = [variance_inflation_factor(data_dummies.values, i) for i in range(data_dummies.shape[1])]

vif_data.sort_values(by="VIF", ascending=False, inplace=True)
vif_data.head(10)  # 상위 10개의 데이터셋
```
